```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns

# X, y 데이터는 이미 정의되어 있다고 가정

li = [i for i in range(10, 500, 20)] # 간격을 20으로 넓혀서 속도 향상
train_scores = [] # 학습 점수 저장용
val_scores = []   # 검증 점수 저장용

for i in li:
    # 1. 모델 생성 (random_state 고정)
    xgb = XGBClassifier(n_estimators=i, random_state=42, eval_metric='logloss')

    # 2. 학습
    xgb.fit(X_train, y_train)

    # 3. 학습 데이터 평가 (과적합 확인용)
    y_train_proba = xgb.predict_proba(X_train)[:, 1]
    train_auc = roc_auc_score(y_train, y_train_proba)
    train_scores.append(train_auc)

    # 4. 검증 데이터 평가 (실제 성능용)
    y_val_proba = xgb.predict_proba(X_val)[:, 1]
    val_auc = roc_auc_score(y_val, y_val_proba)
    val_scores.append(val_auc)

    # 진행 상황 출력 (선택)
    # print(f"n_estimators={i} | Train: {train_auc:.4f} | Val: {val_auc:.4f}")

# 5. 그래프 그리기 (두 선을 함께 표시)
plt.figure(figsize=(12, 6))
sns.lineplot(x=li, y=train_scores, label='Train Score (Training)', marker='o', linestyle='--')
sns.lineplot(x=li, y=val_scores, label='Val Score (Testing)', marker='o')

plt.title("n_estimators vs ROC-AUC Score")
plt.xlabel("n_estimators")
plt.ylabel("ROC-AUC")
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt # ★ 시각화를 위한 라이브러리 추가
from sklearn.model_selection import train_test_split, GridSearchCV
from xgboost import XGBClassifier
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    classification_report,
    roc_auc_score,
    f1_score
)

# (가정) X, y 데이터가 이미 준비되어 있다고 가정합니다.
# X, y = ...

# =========================================================
# 1. 데이터 2단계 분할 (Double Split)
# =========================================================

# [1차 분할] 전체 데이터 -> Final Test Set (20%) / Training Set (80%)
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, test_size=0.2, random_state=19, stratify=y
)

# [2차 분할] Training Set -> Real Train (80%) / Validation (20%)
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.2, random_state=19, stratify=y_temp
)

print(f"전체 데이터 크기: {X.shape}")
print(f"학습용(Train): {X_train.shape}, 검증용(Valid): {X_val.shape}, 평가용(Test): {X_test.shape}")

# =========================================================
# 2. 모델 및 파라미터 그리드 설정
# =========================================================

xgb = XGBClassifier(
    n_estimators=1000,
    eval_metric='logloss', # 시각화할 지표 (Log Loss)
    random_state=42,
    n_jobs=-1,
    early_stopping_rounds=100 # 생성자 혹은 fit에 넣을 수 있음
)

param_grid = {
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 5, 7],
    'scale_pos_weight': [1, 2]
}

# =========================================================
# 3. GridSearchCV 설정
# =========================================================

grid_search = GridSearchCV(
    estimator=xgb,
    param_grid=param_grid,
    cv=3,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=1,
    refit=True
)

# =========================================================
# 4. 학습 수행 (Early Stopping 적용)
# =========================================================

print("GridSearchCV 수행 중... (최적 파라미터 탐색)")
grid_search.fit(
    X_train, y_train,
    # eval_set의 0번은 Train, 1번은 Validation으로 기록됩니다.
    eval_set=[(X_train, y_train), (X_val, y_val)],
    verbose=False
)

# 5. 최적의 모델 추출
best_model = grid_search.best_estimator_

# =========================================================
# ★ [추가된 부분] 학습 곡선(Learning Curve) 시각화 ★
# =========================================================

# best_model에서 학습 이력(logloss 변화)을 가져옵니다.
results = best_model.evals_result()

# eval_set에 넣은 순서대로 키가 생성됩니다.
# validation_0 : (X_train, y_train) -> 학습 데이터 점수
# validation_1 : (X_val, y_val)     -> 검증 데이터 점수
train_loss = results['validation_0']['logloss']
val_loss = results['validation_1']['logloss']
epochs = len(train_loss)
x_axis = range(0, epochs)

# 그래프 그리기
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(x_axis, train_loss, label='Train Loss')
ax.plot(x_axis, val_loss, label='Validation Loss', linestyle='--')
ax.legend()
plt.ylabel('Log Loss')
plt.xlabel('Epochs (Iterations)')
plt.title('XGBoost Learning Curve (Log Loss)')
plt.grid(True)
plt.show() # 그래프 출력

# =========================================================
# 6. 최종 평가 (Test Set 사용)
# =========================================================

y_pred = best_model.predict(X_test)
y_pred_proba = best_model.predict_proba(X_test)[:, 1]

print("\n" + "="*40)
print(f"최적 파라미터: {grid_search.best_params_}")
if hasattr(best_model, 'best_iteration'):
    print(f"최적 반복 횟수(Best Iteration): {best_model.best_iteration}")
print("="*40)

# 1. 기본 지표 출력
print(f"Accuracy (정확도): {accuracy_score(y_test, y_pred):.4f}")
print(f"ROC-AUC Score   : {roc_auc_score(y_test, y_pred_proba):.4f}")

print("-" * 40)

# 2. 혼동 행렬 (Confusion Matrix)
print("Confusion Matrix (혼동 행렬):")
print(confusion_matrix(y_test, y_pred))

print("-" * 40)

# 3. 상세 리포트
print("Classification Report (상세 보고서):")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

# 1. 확률값 예측
y_pred_proba = best_model.predict_proba(X_val)[:, 1]

# 2. PR Curve 계산
precisions, recalls, thresholds = precision_recall_curve(y_val, y_pred_proba)

# 3. 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(thresholds, precisions[:-1], label="Precision")
plt.plot(thresholds, recalls[:-1], label="Recall")
plt.xlabel("Threshold")
plt.legend()
plt.title("Precision-Recall Curve")
plt.grid()
plt.show()
```

```python
# 1. 확률값(Probability) 추출
# [:, 1]은 클래스 1(Positive)일 확률만 가져옵니다.
y_pred_proba = best_model.predict_proba(X_test)[:, 1]

# ---------------------------------------------------------
# ★ 사용자가 원하는 임계값 설정 ★
# ---------------------------------------------------------
CUSTOM_THRESHOLD = 0.4  # 예: 0.35 (35% 이상이면 1로 판단)

# 2. 임계값 적용하여 0과 1로 변환
# (확률 >= 임계값)이면 True(1), 아니면 False(0) -> int로 변환
y_pred_custom = (y_pred_proba >= CUSTOM_THRESHOLD).astype(int)

print("\n" + "="*40)
print(f"최적 파라미터: {grid_search.best_params_}")
print(f"★ 적용된 임계값 (Threshold): {CUSTOM_THRESHOLD}")
print("="*40)

# 3. 사용자 지정 임계값 기준 지표 출력
print(f"Accuracy (정확도): {accuracy_score(y_test, y_pred_custom):.4f}")
print(f"F1 Score (F1점수): {f1_score(y_test, y_pred_custom):.4f}")
print(f"ROC-AUC Score   : {roc_auc_score(y_test, y_pred_proba):.4f}") # AUC는 임계값 무관

print("-" * 40)

# 4. 혼동 행렬 (Confusion Matrix)
print("Confusion Matrix (Custom Threshold):")
print(confusion_matrix(y_test, y_pred_custom))

print("-" * 40)

# 5. 상세 리포트
print("Classification Report (Custom Threshold):")
print(classification_report(y_test, y_pred_custom))
```
